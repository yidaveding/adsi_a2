```python
# nominate columns
col_tar = ['beer_style']
col_cat = ['brewery_name']
col_num = ['review_aroma', 'review_appearance', 'review_palate', 'review_taste']
```

```python
filepath = 'data/processed/beer_review_cleaned.csv'

# load data from csv
df = pd.read_csv('data/raw/beer_reviews.csv')

# clean up dataset: drop unrelated columns and drop rows that contain NA 
df_cleaned = df.copy()
df_cleaned = df[col_num + col_tar + col_cat]

# store data
df_cleaned.dropna(inplace=True)
# df_cleaned.to_csv('data/processed/beer_review_cleaned.csv', index=False)
```

```python
df = df_cleaned['beer_style'].value_counts()
```

```python
df = pd.DataFrame(df)

df['pct'] = round(df['beer_style']/df.beer_style.sum()*100,1)
df.head(10)
```

```python
same_scores = df_cleaned.groupby(['review_aroma', 'review_appearance', 'review_palate', 'review_taste', 'brewery_name', 'beer_style'])\
                .size().reset_index().rename(columns={0:'count'})
same_scores = same_scores.groupby(['review_aroma', 'review_appearance', 'review_palate', 'review_taste', 'brewery_name'])\
                .size().reset_index().rename(columns={0:'unique styles'})

same_scores.sort_values('unique styles', ascending=False, inplace=True)

same_scores.head(10)
```
